```python
import requests 
from bs4 import BeautifulSoup
import time
for index, row in sacred_texts_df[1374:].iterrows():                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                 
  contents = """"""
  title = row.title
  URL = row.url
  # print(URL, title)
  page = requests.get(URL)
  book_soup = BeautifulSoup(page.content, "html.parser")
  results = book_soup.find("body", recursive=True)
  if results:
    content_block = results.find("div", class_ ="w-full block content", recursive=True)
    if content_block:
      results = content_block
  else: 
    results = book_soup.find("table")
  try:
    links = results.find_all("a")
    #print(links)
  except AttributeError:
    print("no links in child component")
  else:
    links = results.find_all("a", recursive=True)
    if "Title Page" not in [link.text for link in links]:
      contents += results.text.strip()
      with open(f'full_texts/{title}.txt', 'w') as f: 
        f.write(contents)
      time.sleep(10)
    else:
      condition = False
      for link in links:  
        if link.text == "Title Page":
          condition = True
        if condition == True:
            try:
              link["href"]
            except KeyError:
              print("bad link")
            else:
              ext = link["href"]
              if "https" in ext:
                part_link = ext 
              else: 
                part_link = sacred_texts_df.url[index][:-9] + ext
            part_page = requests.get(part_link)
            part_soup = BeautifulSoup(part_page.content, "html.parser")
            results = part_soup.find("body") 
            if results:
              print(f"Scraping from: {part_link}")
              body_text = results.text.strip()
            else:
              results = part_soup.find_all(attrs={"align" : "center"})
              body_text = """"""
              item_text = [ item.text for item in results]
              print(f"Scraping from: {part_link}")
              for item in item_text:
                body_text += item + "\n"
              #print(body_text)
            contents += body_text
    with open(f'full_texts/{title}.txt', 'w') as f: 
      f.write(contents)
    time.sleep(10)
```
