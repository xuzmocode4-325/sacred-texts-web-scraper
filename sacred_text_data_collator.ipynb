# Sacred Text Website Data Collator Notebook

```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://sacred-texts.com"
page = requests.get(URL)
sacred_soup = BeautifulSoup(page.content, "html.parser")
results = sacred_soup.find("div",  class_="left-menu hidden md:block")
print(results.prettify())
```

```python
categories = results.find_all("div", class_ = "page-menu", recursive=False)
print(len(categories))
```

```python
catagory_pages = []
```

```python
for category in categories:
    link = category.find("a", class_="underline hover:text-indigo-600")
    link_title = link.text
    link_url = link["href"]
    cat_obj = {'category': link_title, 'url':link_url}
    catagory_pages.append(cat_obj)
```

```python
catagory_pages[0]
```

```python
afr_page = catagory_pages[0]
afr_url = afr_page['url']
afr_page_dl = requests.get(afr_url)
afr_soup = BeautifulSoup(afr_page_dl.content, "html.parser")
afr_results = afr_soup.find("div",  class_="w-full block content")
print(afr_results.prettify())
```

```python
afr_books = afr_results.find_all("span", class_="c_e")
```

```python
print(len(afr_books))
```

```python
text_placeholder = lambda soup_item: soup_item.text if soup_item else "none provided" 
url_placeholder = lambda soup_item: soup_item["href"] if soup_item else "none provided"
```

```python
for book in afr_books:
    link = book.find("a")
    author_span = book.find("span", class_="c_a")
    book_obj = {'title': text_placeholder(link), 'author': text_placeholder(author_span), 'url':url_placeholder(link)}
    #catagory_pages[0]["book list"].append(book_obj)
    print(book_obj, end='\n')
    #print(link_url, link.text, author.text)
    #break
    #link_title = link.text
    #link_url = link["href"]
    #cat_obj = {'category': link_title, 'url':link_url}
    #catagory_pages.append(cat_obj)
    #print(link_title, link_url, end='\n')
```

```python
for cat in catagory_pages:
    cat["books"] = []
    cat_url = cat["url"]
    cat_page_dl = requests.get(cat_url)
    cat_soup = BeautifulSoup(cat_page_dl.content, "html.parser")
    cat_results = cat_soup.find("div",  class_="w-full block content")
    cat_books = cat_results.find_all("span", class_="c_e")
    print(f"scraping {cat['category']}")
    for book in cat_books:
        link = book.find("a")
        author_span = book.find("span", class_="c_a")
        book_obj = {'title': text_placeholder(link), 'author': text_placeholder(author_span), 'url':url_placeholder(link)}
        cat["books"].append(book_obj)
```

```python
catagory_pages[3]["books"]
```

```python
sacred_books = []
for cat in catagory_pages:
    for book in cat["books"]:
        book["category"] = cat["category"]
        sacred_books.append(book)
```

```python
import pandas as pd
sacred_books_df = pd.DataFrame(sacred_books)
sacred_books_df.head()
```

```python
id = [sacred_books_df.url[row][29:-10] for row in sacred_books_df.index]
sacred_books_df["id"] = id  
```

```python
dataset = pd.DataFrame.to_csv(sacred_books_df)
```
